```python
import pandas as pd
import numpy as np
import pymongo
import json
```

```python
df_movies = pd.read_csv('movies.csv', encoding = "utf-8")
df_movies = df_movies.replace(np.nan, '', regex=True)
df_movies.head()
```

```python
df_movies.tail()
```

```python
titles = df_movies['title'].tolist()
len(titles)
```

```python
names=[]
years=[]
sps=[]
```

```python
for title in titles:
    sp=title.split('(')
    sps.append(sp)
```

```python
for s in sps:
    if len(s)!=2:
        print(s)
```

```python
len(sps[120])
```

```python
for s in sps:
    if len(s)==1:
        print(s)
```

```python
for s in sps:
    names.append(s[0])
```

```python
len(names)
```

```python
n=[]
for name in names:
    x=name.split(',')
    y=x[0]
    n.append(y.strip())
```

```python
len(n)
```

```python
df_movies.drop(columns=['title'], inplace=True)
```

```python
df_movies['title']=n
```

```python
df_movies.head()
```

```python
df_rate = pd.read_csv('ratings.csv', encoding='utf-8')
df_rate.fillna(0, inplace=True)
df_rate.drop(columns=['timestamp'], inplace=True)
df_rate.head()
```

```python
df_merge = pd.merge(df_movies, df_rate, on='movieId')
df_merge.head()
```

```python
df_merge = df_merge[['userId', 'movieId', 'title', 'rating']]
df_merge.head()
```

```python
df_merge.shape

df_merge.isnull().values.any()
```

```python
df_merge.shape
```

```python
df_merge_json = json.loads(df_merge.T.to_json()).values()
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient()
```

```python
db = client.hollywoodDB
```

```python
db.merge.insert_many(df_merge_json)
```
